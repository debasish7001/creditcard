This is anunbalanced Data set. So we have to balanced,here we used oversampling to balanced.

Balanced Dataset

```python
from sklearn.utils import resample
```

```python
majority=df[(df['Class']==0)]
```

```python
mainority=df[(df['Class']==1)]
```

```python
df1=resample(mainority,replace=True,n_samples=284315,random_state=42)
```

```python
df2=pd.concat([df1,majority])
```

```python
sns.countplot(df2['Class'])
```

# Input Split

```python
x = df2.drop(columns=['Class'], axis=1)
y = df2['Class']
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x1= sc.fit_transform(x)
```

# Model Training

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x1,y,train_size=0.6)
```

# Logestic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='newton-cg')
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
#from sklearn.metrics import confusion_matrix
#confusion_matrix(lr.predict(x_test),y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

# xgboost

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
xgboost = GradientBoostingClassifier()
```

```python
xgboost.fit(x_train,y_train)
```

```python
xgboost.score(x_test,y_test)
```

Logestic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(solver='newton-cg')
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report, f1_score
```

```python
y_pred = lr.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```
